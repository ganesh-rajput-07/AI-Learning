# Exercise 10 — AI Dataset Visualization ⭐⭐⭐⭐

Generate

```python
1000 students
```

Columns

```text
Maths

Physics

Chemistry

Attendance
```

Now visualize

* Histogram of Maths
* Histogram of Physics
* Histogram of Chemistry
* Scatter Plot (Maths vs Physics)
* Scatter Plot (Attendance vs Maths)
* Bar Chart (Top 10 Students)
* Pie Chart (Pass/Fail)

Finally answer

1. Which subject has highest average?

2. Does attendance affect marks?

3. Which graph gives the most useful information?


````python
import numpy as np
import matplotlib.pyplot as plt

# data_x = np.array(range(1,1001))


maths = np.array(np.random.normal(0,101,1000))
physics = np.array(np.random.normal(0,101,1000))
chemistry = np.array(np.random.normal(0,101,1000))
attendance = np.array(np.random.randint(0,225,1000))

plt.hist(maths)
plt.show()
plt.hist(physics)
plt.show()
plt.hist(chemistry)

plt.show()

plt.scatter(maths, physics)
plt.show()
plt.scatter(attendance, maths)
plt.show()

total = maths + physics + chemistry
top10 = np.argsort(total)[-10:]
plt.bar(range(10), total[top10])

plt.xticks(range(10), top10)

plt.title("Top 10 Students")

plt.xlabel("Student ID")

plt.ylabel("Total Marks")

plt.show()

average = total / 3

passed = np.sum(average >= 40)



failed = np.sum(average < 40)

plt.pie(

    [passed, failed],

    labels=["Pass", "Fail"],

    autopct="%1.1f%%"

)



plt.title("Pass vs Fail")



plt.show()
````
