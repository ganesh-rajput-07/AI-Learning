# Level 6 — Boolean Indexing

Create

```python
arr = np.arange(1,51)
```

Find

* Numbers greater than 25
* Even numbers
* Odd numbers
* Multiples of 5
* Numbers between 10 and 30
* Numbers divisible by both 2 and 3

Do not use loops.


````python
arr = np.arange(1,51)

print("Numbers greater than 25 : ", arr[arr>25])
print("Even numbers : ", arr[arr%2==0])
print("Odd numbers : ",arr[arr%2!=0])
print("Multiples of 5 : ", arr[arr%5 == 0])
print("Numbers between 10 and 30 : ", arr[(arr>=10) & (arr<=30)])
print("Numbers divisible by both 2 and 3 : ", arr[(arr%2 == 0) & (arr%3 == 0)])
````

# Final Challenge 🏆

Without watching any tutorial, build a small **Student Analytics System** using only NumPy.

The program should:

* Generate marks for 500 students.
* Calculate average, highest, and lowest marks.
* Count passed and failed students.
* Find the top 10 students.
* Normalize marks.
* Standardize marks.
* Display grade distribution (A, B, C, D, F).
* Calculate the pass percentage.conditions = [
    marks >= 90,
    marks >= 80,
    marks >= 70,
    marks >= 50
]

choices = ["A", "B", "C", "D"]

grades = np.select(conditions, choices, default="F")

* Print all results without using Python loops wherever possible.

---


````python
## 1 Generate marks for 500 students.

marks = random.randint(0,101,500)

print("Marks : ", marks)

## 2 Calculate average, highest, and lowest marks

print("Average of marks : ", np.average(marks))

print("Highest Marks : ", np.sort(marks)[-1])

print("Lowest Marks : ", np.sort(marks)[1])

## 3 Count passed and failed students.

print("Passed students : ", np.sum(marks>=36))

print("Failed students : ", np.sum(marks<36))

print("The top 10 students : ", np.sort(marks)[-10:][-1])

print("Normalized Marks : ", (marks - np.min(marks))/(np.max(marks)-np.min(marks)))

print("Standardize marks : ", (marks - np.mean(marks))/(np.std(marks)))


conditions = [
    marks >= 90,
    marks >= 80,
    marks >= 70,
    marks >= 50
]

choices = ["A", "B", "C", "D"]

grades = np.select(conditions, choices, default="F")

print("grade distribution : ", grades)

print("The pass percentage : ", np.sum(marks>=36)*100/500)
````
